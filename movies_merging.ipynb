```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hollywood_df = pd.read_csv('Datasets/modified_hollywood_movies.csv') #loading the both modified datasets
indian_df = pd.read_csv('Datasets/modified_indian_movies.csv')
```

```python
final_movies_df = pd.concat([hollywood_df, indian_df], axis=0).reset_index(drop=True) #merging the datasets
final_movies_df
```

```python
final_movies_df[final_movies_df.duplicated()] #checking for duplicate rows after merging
```

```python
final_movies_df[final_movies_df.duplicated(subset=['Name'])] #checking for duplicate movie names after merging
```

```python
final_movies_df[final_movies_df.duplicated(subset=['Name','Year','Language'])].shape #checking for duplicate movie names along with year and language 
```

```python
final_movies_df.drop_duplicates(subset=['Name','Year','Language'], keep='first', inplace=True) #dropping duplicate movie names along with year and language
```

```python
final_movies_df.reset_index(drop=True, inplace=True)
final_movies_df
```

```python
final_movies_df.duplicated().any() #rechecking for duplicate rows
```

```python
final_movies_df = final_movies_df.sort_values(by='Year', ascending= True).reset_index(drop=True)
final_movies_df
```

```python
final_movies_df['Year'] = final_movies_df['Year'].astype('Int64') #converting year column to integer type
```

```python
final_movies_df.dtypes
```

```python
final_movies_df.to_csv('Datasets/final_movies_dataset.csv', index=False) #saving the final merged dataset for visualization and analysis
```
